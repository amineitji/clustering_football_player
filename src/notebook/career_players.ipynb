# SCRAPING DATA

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

# Créer le dossier ../../data/comparaison/ si nécessaire
output_dir = "../../data/comparaison/"
os.makedirs(output_dir, exist_ok=True)

# Fonction pour scraper les tableaux spécifiques d'une page web et les sauvegarder dans un fichier CSV
def scrape_and_save(url, player_name):
    # Envoyer une requête à la page web
    response = requests.get(url)
    
    if response.status_code == 200:
        # Parse la page avec BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Liste des IDs des tableaux à scraper
        table_ids = [
            'stats_passing_expanded', 
            'stats_standard_expanded', 
            'stats_shooting_expanded',  
            'stats_gca_expanded', 
            'stats_defense_expanded',
            'stats_possession_expanded'
        ]
        
        # Scraper et sauvegarder uniquement les tableaux avec les IDs spécifiés
        for table_id in table_ids:
            table = soup.find('table', {'id': table_id})
            if table:
                # Lire le tableau avec pandas
                df = pd.read_html(str(table))[0]
                
                # Définir le nom du fichier CSV
                csv_filename = os.path.join(output_dir, f"{player_name}_{table_id}.csv")
                
                # Sauvegarder le dataframe dans un fichier CSV
                df.to_csv(csv_filename, index=False)
                print(f"Tableau {table_id} pour {player_name} sauvegardé dans {csv_filename}")
            else:
                print(f"Tableau avec l'ID {table_id} non trouvé pour {player_name}")
    else:
        print(f"Échec de la requête pour {url}. Code de statut : {response.status_code}")

# URLs des pages à scraper
urls = [
    ("https://fbref.com/fr/joueurs/df69b544/all_comps/Statistiques-Antoine-Griezmann-Stats---Toutes-les-competitions", "Griezmann"),
    ("https://fbref.com/fr/joueurs/69384e5d/all_comps/Statistiques-Neymar-Stats---Toutes-les-competitions", "Neymar")
]

# Scraper chaque page et sauvegarder les tableaux
for url, player_name in urls:
    scrape_and_save(url, player_name)
```

```python
import pandas as pd
import os
import re

# Fonction pour nettoyer et transformer les données d'un fichier CSV
def clean_and_transform_csv(file_path):
    # Lire le fichier CSV en ignorant la première ligne
    df = pd.read_csv(file_path, skiprows=1)
    
    # Définir une expression régulière pour détecter les lignes qui commencent par une année au format 'yyyy-yyyy'
    date_pattern = re.compile(r'^\d{4}-\d{4}$')
    
    # Filtrer les lignes qui commencent par une date
    df_cleaned = df[df.iloc[:, 0].astype(str).apply(lambda x: bool(date_pattern.match(x)))]
    
    # Remplacer les valeurs vides (NaN) et les chaînes vides par None
    df_cleaned = df_cleaned.replace("", None).where(pd.notnull(df_cleaned), None)

    
    # Supprimer la colonne "Matchs" si elle existe
    if "Matchs" in df_cleaned.columns:
        df_cleaned = df_cleaned.drop(columns=["Matchs"])
    
    # Réécrire le fichier modifié
    df_cleaned.to_csv(file_path, index=False)
    print(f"Fichier nettoyé et transformé : {file_path}")

# Dossier contenant les CSV à nettoyer
csv_dir = "../../data/comparaison/"

# Nettoyer et transformer tous les fichiers CSV dans le dossier
for file_name in os.listdir(csv_dir):
    if file_name.endswith(".csv"):
        file_path = os.path.join(csv_dir, file_name)
        clean_and_transform_csv(file_path)
```

# Unification des données

```python
import pandas as pd
import os

def merge_player_stats_and_delete_files(player_name, file_paths, output_dir="../../data/comparaison/"):
    """
    Cette fonction prend un nom de joueur, une liste de chemins de fichiers CSV associés à ce joueur,
    effectue une jointure des fichiers CSV, fusionne les colonnes identiques avec des données similaires,
    enregistre le résultat final dans un fichier CSV et supprime les fichiers CSV originaux.

    :param player_name: Nom du joueur (ex: 'Griezmann', 'Neymar')
    :param file_paths: Liste des chemins des fichiers CSV à fusionner
    :param output_dir: Répertoire où sauvegarder les fichiers fusionnés
    :return: Chemin du fichier CSV fusionné
    """
    # Créer le répertoire de sortie s'il n'existe pas
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Charger tous les CSV
    dfs = [pd.read_csv(file) for file in file_paths]

    # Colonnes communes pour la jointure
    common_columns = ['Saison', 'Âge', 'Équipe', 'Comp']

    # Fusionner tous les DataFrames
    df_merged = dfs[0]
    for df in dfs[1:]:
        df_merged = df_merged.merge(df, on=common_columns, how='outer')

    # Fusionner les colonnes identiques en vérifiant si elles ont des valeurs identiques
    df_final = df_merged.loc[:, ~df_merged.columns.duplicated()]

    for col in df_merged.columns[df_merged.columns.duplicated(keep=False)]:
        duplicated_columns = df_merged.loc[:, df_merged.columns == col]
        if (duplicated_columns.nunique(axis=1) == 1).all():  # Si les colonnes sont identiques
            df_final[col] = duplicated_columns.iloc[:, 0]  # Garde une seule colonne
        else:
            # Garde toutes les colonnes si elles contiennent des valeurs différentes
            df_final = pd.concat([df_final, duplicated_columns], axis=1)

    # Sauvegarder le fichier fusionné
    output_file = os.path.join(output_dir, f"{player_name}_merged_stats.csv")
    df_final.to_csv(output_file, index=False)

    # Supprimer les fichiers CSV originaux
    for file in file_paths:
        os.remove(file)

    print(f"Tous les fichiers CSV ont été supprimés après la fusion pour {player_name}.")
    
    return output_file

# Exemple d'utilisation pour Griezmann
griezmann_files = [
    '../../data/comparaison/Griezmann_stats_standard_expanded.csv',
    '../../data/comparaison/Griezmann_stats_shooting_expanded.csv',
    '../../data/comparaison/Griezmann_stats_possession_expanded.csv',
    '../../data/comparaison/Griezmann_stats_passing_expanded.csv',
    '../../data/comparaison/Griezmann_stats_gca_expanded.csv',
    '../../data/comparaison/Griezmann_stats_defense_expanded.csv'
]

# Exemple d'utilisation pour Neymar
neymar_files = [
    '../../data/comparaison/Neymar_stats_standard_expanded.csv',
    '../../data/comparaison/Neymar_stats_shooting_expanded.csv',
    '../../data/comparaison/Neymar_stats_possession_expanded.csv',
    '../../data/comparaison/Neymar_stats_passing_expanded.csv',
    '../../data/comparaison/Neymar_stats_gca_expanded.csv',
    '../../data/comparaison/Neymar_stats_defense_expanded.csv'
]

# Fusionner les fichiers de Griezmann et supprimer les fichiers originaux
griezmann_output = merge_player_stats_and_delete_files('Griezmann', griezmann_files)
print(f"Fichier fusionné pour Griezmann : {griezmann_output}")

# Fusionner les fichiers de Neymar et supprimer les fichiers originaux
neymar_output = merge_player_stats_and_delete_files('Neymar', neymar_files)
print(f"Fichier fusionné pour Neymar : {neymar_output}")
```

# PCA

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
from math import pi

# Charger les fichiers CSV
file_neymar = '../../data/comparaison/Neymar_merged_stats.csv'
file_griezmann = '../../data/comparaison/Griezmann_merged_stats.csv'

neymar_stats = pd.read_csv(file_neymar)
griezmann_stats = pd.read_csv(file_griezmann)

# Fusionner les datasets
data = pd.concat([neymar_stats, griezmann_stats])

# Filtrer pour garder uniquement les compétitions : '1. Champions Lg', '1. La Liga', '1. Ligue 1'
competitions_to_keep = ['1. Champions Lg', '1. La Liga', '1. Ligue 1']
data = data[data['Comp'].isin(competitions_to_keep)]

# Filtrer les saisons à partir de 2017-2018
data = data[data['Saison'] >= '2017-2018']

# Exclure les colonnes non pertinentes
exclude_columns = ['Saison', 'Âge', 'Équipe', 'Comp', 'Titulaire', 'Min']
data_cleaned = data.drop(columns=exclude_columns)

# Remplacer les valeurs -1.0 par NaN avant de fusionner les lignes
data_cleaned = data_cleaned.replace(-1.0, pd.NA)
data_cleaned = data_cleaned.fillna(data_cleaned.mean())

# Fusionner les lignes de la même saison
data_grouped = data.groupby(['Saison', 'Âge', 'Équipe']).sum().reset_index()

# Organiser les catégories en trois composantes principales
finition_columns = ['Buts_x', 'B-PénM', 'PénM_x', 'PénT_x', 'PrgR', 'Buts_y', 'Tirs_x', 'TC', 'TC%', 
                    'Tir/90', 'TC/90', 'B/Tir', 'B/TC', 'Dist', 'CF', 'PénM_y', 'PénT_y', 'Tirs_y']
creation_columns = ['PrgC_x', 'Touches', 'SurfRépDéf', 'ZDéf_x', 'MilTer_x', 'ZOff_x', 'SurfRépOff',
                    'Action de jeu', 'Balle au pied', 'TotDist_x', 'DistBut_x', 'PrgC_y', '1/3_x', 'AMT', 'AMB', 
                    'PD_x', 'PrgP_x', 'Cmp', 'Att_y', 'Cmp%', 'TotDist_y', 'DistBut_y', 'PD_y', 'xAG', 'xA',
                    'A-xAG', 'PC', '1/3_y', 'PPA', 'CntSR', 'PrgP_y', 'PassLive', 'PassDead', 'Att_x', 'Succ', 
                    'Succ%', 'Tkld', 'Tkld%', 'TO', 'Ftp']
defense_columns = ['Déf', 'Tcl', 'TclR', 'ZDéf_y', 'MilTer_y', 'ZOff_y', 'Tcl.1', 'Att', 'Tcl%', 
                   'Balles contrées', 'Tirs', 'Passe', 'Int', 'Tcl+Int', 'Dég']

# Séparer les données en trois groupes pour le PCA
finition_data = data_grouped[finition_columns]
creation_data = data_grouped[creation_columns]
defense_data = data_grouped[defense_columns]

# Normaliser les données pour chaque composante
scaler = StandardScaler()
finition_scaled = scaler.fit_transform(finition_data)
creation_scaled = scaler.fit_transform(creation_data)
defense_scaled = scaler.fit_transform(defense_data)

# Appliquer PCA pour chaque composante
pca_finition = PCA()
pca_creation = PCA()
pca_defense = PCA()

finition_pca_result = pca_finition.fit_transform(finition_scaled)
creation_pca_result = pca_creation.fit_transform(creation_scaled)
defense_pca_result = pca_defense.fit_transform(defense_scaled)

# Visualiser la variance expliquée par chaque composante pour chaque groupe
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(pca_finition.explained_variance_ratio_)+1), pca_finition.explained_variance_ratio_, alpha=0.7, align='center')
plt.ylabel('Variance expliquée')
plt.xlabel('Composante principale')
plt.title('Variance expliquée pour la composante Finition')
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(range(1, len(pca_creation.explained_variance_ratio_)+1), pca_creation.explained_variance_ratio_, alpha=0.7, align='center')
plt.ylabel('Variance expliquée')
plt.xlabel('Composante principale')
plt.title('Variance expliquée pour la composante Création de jeu')
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(range(1, len(pca_defense.explained_variance_ratio_)+1), pca_defense.explained_variance_ratio_, alpha=0.7, align='center')
plt.ylabel('Variance expliquée')
plt.xlabel('Composante principale')
plt.title('Variance expliquée pour la composante Défense')
plt.show()

# Afficher les composantes principales influencées par chaque variable
pca_finition_components = pd.DataFrame(pca_finition.components_, columns=finition_columns)
pca_creation_components = pd.DataFrame(pca_creation.components_, columns=creation_columns)
pca_defense_components = pd.DataFrame(pca_defense.components_, columns=defense_columns)

print("PCA Finition Composantes :")
print(pca_finition_components)

print("PCA Création de jeu Composantes :")
print(pca_creation_components)

print("PCA Défense Composantes :")
print(pca_defense_components)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import numpy as np
from math import pi
import matplotlib.colors as mcolors
from matplotlib.gridspec import GridSpec

# Charger les fichiers CSV
file_neymar = '../../data/comparaison/Neymar_merged_stats.csv'
file_griezmann = '../../data/comparaison/Griezmann_merged_stats.csv'

neymar_stats = pd.read_csv(file_neymar)
griezmann_stats = pd.read_csv(file_griezmann)

# Ajouter une colonne 'Joueur' pour différencier Neymar et Griezmann
neymar_stats['Joueur'] = 'Neymar'
griezmann_stats['Joueur'] = 'Griezmann'

# Fusionner les datasets
data = pd.concat([neymar_stats, griezmann_stats])

# Filtrer pour garder uniquement les compétitions : '1. Champions Lg', '1. La Liga', '1. Ligue 1'
competitions_to_keep = ['1. Champions Lg', '1. La Liga', '1. Ligue 1']
data = data[data['Comp'].isin(competitions_to_keep)]

# Filtrer les saisons à partir de 2017-2018
data = data[data['Saison'] >= '2017-2018']

# Exclure les colonnes non pertinentes
exclude_columns = ['Saison', 'Âge', 'Équipe', 'Comp', 'Titulaire', 'Min', 'Joueur']
data_cleaned = data.drop(columns=exclude_columns)

# Sélectionner uniquement les colonnes numériques pour remplir les NaN
numeric_columns = data_cleaned.select_dtypes(include=[np.number]).columns

# Remplacer les valeurs NaN uniquement dans les colonnes numériques
data_cleaned[numeric_columns] = data_cleaned[numeric_columns].fillna(data_cleaned[numeric_columns].mean())

# Fusionner les lignes de la même saison
data_grouped = data.groupby(['Saison', 'Âge', 'Équipe', 'Joueur']).sum().reset_index()

# Organiser les catégories en cinq composantes principales
finition_columns = ['Buts_x', 'B-PénM', 'PénM_x', 'PénT_x', 'PrgR', 'Buts_y', 'Tirs_x', 'TC', 'TC%', 
                    'Tir/90', 'TC/90', 'B/Tir', 'B/TC', 'Dist', 'CF', 'PénM_y', 'PénT_y', 'Tirs_y']
creation_columns = ['PrgC_x', 'Touches', 'SurfRépDéf', 'ZDéf_x', 'MilTer_x', 'ZOff_x', 'SurfRépOff',
                    'Action de jeu', 'Balle au pied', 'TotDist_x', 'DistBut_x', 'PrgC_y', '1/3_x', 'AMT', 'AMB']
playmaker_columns = ['PD_x', 'PrgP_x', 'Cmp', 'Att_y', 'Cmp%', 'TotDist_y', 'DistBut_y', 'PD_y', 'xAG', 'xA',
                     'A-xAG', 'PC', '1/3_y', 'PPA', 'CntSR', 'PrgP_y', 'PassLive', 'PassDead']
dribble_columns = ['Att_x', 'Succ', 'Succ%', 'Tkld', 'Tkld%', 'TO', 'Ftp']
defense_columns = ['Déf', 'Tcl', 'TclR', 'ZDéf_y', 'MilTer_y', 'ZOff_y', 'Tcl.1', 'Att', 'Tcl%', 
                   'Balles contrées', 'Tirs', 'Passe', 'Int', 'Tcl+Int', 'Dég']

# Séparer les données en cinq groupes
finition_data = data_grouped[finition_columns]
creation_data = data_grouped[creation_columns]
playmaker_data = data_grouped[playmaker_columns]
dribble_data = data_grouped[dribble_columns]
defense_data = data_grouped[defense_columns]

# Normaliser chaque groupe de données avec StandardScaler
scaler = StandardScaler()

finition_scaled = pd.DataFrame(scaler.fit_transform(finition_data), columns=finition_columns)
creation_scaled = pd.DataFrame(scaler.fit_transform(creation_data), columns=creation_columns)
playmaker_scaled = pd.DataFrame(scaler.fit_transform(playmaker_data), columns=playmaker_columns)
dribble_scaled = pd.DataFrame(scaler.fit_transform(dribble_data), columns=dribble_columns)
defense_scaled = pd.DataFrame(scaler.fit_transform(defense_data), columns=defense_columns)

# Calculer la moyenne des composantes normalisées par saison et équipe
finition_mean = finition_scaled.mean(axis=1)
creation_mean = creation_scaled.mean(axis=1)
playmaker_mean = playmaker_scaled.mean(axis=1)
dribble_mean = dribble_scaled.mean(axis=1)
defense_mean = defense_scaled.mean(axis=1)

# Combiner les moyennes dans un DataFrame
composantes_data = pd.DataFrame({
    'Saison': data_grouped['Saison'],
    'Équipe': data_grouped['Équipe'],
    'Joueur': data_grouped['Joueur'],
    'Finition': finition_mean,
    'Création': creation_mean,
    'Playmaker': playmaker_mean,
    'Dribble': dribble_mean,
    'Défense': defense_mean
})


# Fonction pour créer un fond en dégradé couvrant 100 % de l'espace
def create_gradient_background(fig, color1="#000000", color2="#3b3700"):
    """Crée un fond en dégradé vertical couvrant toute la figure."""
    gradient = np.linspace(0, 1, 256).reshape(-1, 1)
    gradient = np.hstack((gradient, gradient))
    cmap = mcolors.LinearSegmentedColormap.from_list("", [color1, color2])
    ax = fig.add_axes([0, 0, 1, 1], zorder=-1)
    ax.axis('off')
    ax.imshow(gradient, aspect='auto', cmap=cmap, extent=[0, 1, 0, 1])

# Fonction pour personnaliser les axes (par exemple : contours blancs et épais)
def customize_axes(ax):
    """Personnalise les axes avec du blanc et du gras."""
    ax.spines['polar'].set_color('white')  # Contour des axes en blanc
    ax.spines['polar'].set_linewidth(2)  # Contours plus épais

# Fonction pour créer un radar chart avec un style personnalisé
def create_styled_radar_chart(data, categories, saison, category_labels, color1="#000000", color2="#3b3700"):
    N = len(categories)

    # Angles pour chaque catégorie
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]  # Boucler pour fermer le cercle

    # Initialiser la figure avec un GridSpec pour un meilleur contrôle
    fig = plt.figure(figsize=(14, 14),facecolor='none')
    gs = GridSpec(1, 1)  # 1 seul graphique sur toute la figure

    # Créer le fond en dégradé
    create_gradient_background(fig, color1, color2)

    # Colonne de gauche : radar chart
    ax = fig.add_subplot(gs[0], polar=True, facecolor='none')  # Fond transparent

    # Tracer pour chaque joueur
    for i, row in data.iterrows():
        values = row[categories].values.flatten().tolist()
        values += values[:1]  # Boucler pour fermer le cercle

        if row['Joueur'] == 'Neymar':
            # Tracer en mettant en avant Neymar
            ax.plot(angles, values, linewidth=4, linestyle='solid', label='Neymar', color='red', zorder=4)
            ax.fill(angles, values, color='red', alpha=0.3)
        elif row['Joueur'] == 'Griezmann':
            # Tracer en mettant en avant Griezmann
            ax.plot(angles, values, linewidth=4, linestyle='solid', label='Griezmann', color='blue', zorder=4)
            ax.fill(angles, values, color='blue', alpha=0.3)

    # Dessiner les labels personnalisés pour chaque catégorie
    plt.xticks(angles[:-1], category_labels, color='white', fontsize=18, fontweight='bold')

    # Enlever les valeurs numériques sur les axes
    ax.set_yticklabels([])

    # Personnaliser les axes (par exemple : contours blancs et épais)
    customize_axes(ax)

    # Ajouter le titre
    ax.set_title(f'Neymar vs Griezmann - Saison {saison}', size=25, color='white', fontweight='bold', y=1.2)

    # Appliquer la personnalisation des axes (graduations en blanc)
    ax.tick_params(axis='x', colors='white', labelsize=20,pad=20)
    
    # Ajouter l'étiquette Twitter
    ax.text(1.15, 0.94, "@TarbouchData", fontsize=18, color='white', ha='center', transform=ax.transAxes, alpha=0.8,fontweight='bold')

    # Afficher la légende avec du texte en blanc
    legend = ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1), fontsize=20)
    plt.setp(legend.get_texts(), color='black')

    # Afficher le graphique
    plt.tight_layout(pad=1)
    plt.show()


# Définir les catégories
categories = ['Finition', 'Création', 'Playmaker', 'Dribble', 'Défense']
category_labels = ['Qualité de finition', 'Création balle au pied', 'Création à la passe', 'Qualité de Dribbles', 'Efforts défensifs']

# Créer une boucle pour chaque saison
seasons = composantes_data['Saison'].unique()

for saison in seasons:
    # Filtrer les données pour Neymar et Griezmann pour chaque saison
    data_for_season = composantes_data[(composantes_data['Saison'] == saison) & (composantes_data['Joueur'].isin(['Neymar', 'Griezmann']))]
    
    # Créer le radar chart avec le style personnalisé
    create_styled_radar_chart(data_for_season, categories, saison, category_labels)

```
