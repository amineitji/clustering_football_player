```python
import csv
import requests
from bs4 import BeautifulSoup
import time

class PlayerScraper:
    def __init__(self, team_name, team_url):
        self.team_name = team_name + "2020"
        self.team_url = team_url

    def scrape_players(self):
        # Define the output CSV file
        output_file = '../../data/players_data.csv'
        
        # Open the CSV file in append mode
        with open(output_file, mode='a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['Player Name', 'Position', 'Player URL', 'Team Name'])  # Updated header to include Team Name

            print(f"Scraping players from {self.team_name}...")

            # Fetch the page content and parse it
            response = requests.get(self.team_url)
            soup = BeautifulSoup(response.content, 'html.parser')

            # Find the player table
            player_table = soup.find('table', {'class': 'stats_table sortable min_width'})

            # Iterate through player rows
            for player_row in player_table.find('tbody').find_all('tr'):
                # Extract player name from the <th> element
                player_name = player_row.find('th', {'data-stat': 'player'}).get_text(strip=True)

                # Extract player position from the <td> element
                player_position = player_row.find('td', {'data-stat': 'position'}).get_text(strip=True)

                # Extract the player link from the <a> inside the <th> element
                player_link = player_row.find('th', {'data-stat': 'player'}).find('a')['href']
                player_link = 'https://fbref.com' + player_link
                print(player_name, " - ", player_position, "- ", player_link, " - ", self.team_name)

                # Write player data along with team name to the CSV file
                writer.writerow([player_name, player_position, player_link, self.team_name])

        print(f"Players data saved to {output_file}")

# Usage for a single club
url = 'https://fbref.com/en/squads/e2d8892c/2019-2020/Paris-Saint-Germain-Stats'
team_name = 'Paris Saint-Germain'

# Create an instance of PlayerScraper
scraper = PlayerScraper(team_name, url)

# Scrape player data for the team
scraper.scrape_players()
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import os
import pandas as pd
import time

class DataPlayerExtractor:
    def __init__(self, players_data_path, output_csv_path='../../data/scouting_report copy.csv'):
        self.players_data_path = players_data_path
        self.output_csv_path = output_csv_path
        self.soup = None
        self.players_df = None

    def fetch_data(self, url):
        """Fetch the content of the webpage."""
        try:
            response = requests.get(url)
            response.raise_for_status()
            self.soup = BeautifulSoup(response.text, 'html.parser')
        except requests.exceptions.HTTPError as http_err:
            print(f"HTTP error occurred: {http_err}")
        except Exception as err:
            print(f"Other error occurred: {err}")
    
    def extract_player_name(self):
        """Extract the player name from the webpage."""
        if self.soup:
            player_name_tag = self.soup.find('h1')
            return player_name_tag.text.strip() if player_name_tag else "Nom non trouvé"
        return None

    def extract_scouting_report(self):
        """Extract the scouting report (if any) from the webpage."""
        if self.soup:
            table = self.soup.find('table')
            if not table:
                print("Table introuvable.")
                return None, None
            
            headers = []
            rows_data = []
            
            # Extraction des en-têtes de tableau
            thead = table.find('thead')
            if thead:
                headers = [th.text.strip() for th in thead.find_all('th')]

            # Extraction des lignes du tableau
            tbody = table.find('tbody')
            if tbody:
                for row in tbody.find_all('tr'):
                    row_data = []
                    for cell in row.find_all(['th', 'td']):
                        row_data.append(cell.text.strip())
                    
                    # Ignorer les lignes vides
                    if any(row_data):
                        rows_data.append(row_data)
            
            return headers, rows_data
        return None, None

    def save_to_csv(self, player_name, rows_data):
        """Save player data to a CSV file."""
        # Colonnes à utiliser dans le CSV
        columns = ['player_name'] + [row[0] for row in rows_data]

        # Extraire uniquement les valeurs (sans le nom des statistiques)
        values = [player_name] + [row[1] for row in rows_data]

        # Vérifier si le fichier CSV existe déjà
        file_exists = os.path.isfile(self.output_csv_path)

        # Créer le répertoire 'data' si nécessaire
        os.makedirs(os.path.dirname(self.output_csv_path), exist_ok=True)

        # Ajouter les données au CSV
        with open(self.output_csv_path, 'a', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            if not file_exists:
                # Écrire les en-têtes si le fichier est créé pour la première fois
                writer.writerow(columns)
            # Écrire les données du joueur
            writer.writerow(values)
        
        print(f"Données sauvegardées dans {self.output_csv_path}")

    def process_players(self):
        """Process all players listed in the input CSV."""
        # Charger le fichier players_data.csv
        self.players_df = pd.read_csv(self.players_data_path)

        # Initialiser le fichier de sortie scouting_report.csv s'il n'existe pas
        if not os.path.isfile(self.output_csv_path):
            with open(self.output_csv_path, 'w', newline='', encoding='utf-8') as f:
                writer = csv.writer(f)
                writer.writerow(['Player Name', 'Stat Name 1', 'Stat Value 1', 'Stat Name 2', 'Stat Value 2', '...'])

        # Boucle sur chaque joueur et extraction des données
        for index, row in self.players_df.iterrows():
            player_name = row['Player Name']
            player_url = row['Player URL']
            print(f"Processing {player_name} ({player_url})")

            self.fetch_data(player_url)
            headers, scouting_data = self.extract_scouting_report()
            
            if scouting_data:
                self.save_to_csv(player_name, scouting_data)
            else:
                print(f"No scouting report found for {player_name}.")
            
            # Attendre 12 secondes avant de traiter le prochain joueur
            time.sleep(12)

if __name__ == "__main__":
    players_data = [
        {"Player Name": "Ángel Di María", "Player URL": "https://fbref.com/en/players/19cda00b/scout/3243/Angel-Di-Maria-Scouting-Report"},
        {"Player Name": "Keylor Navas", "Player URL": "https://fbref.com/en/players/ecada4fc/scout/3243/Keylor-Navas-Scouting-Report"},
        {"Player Name": "Thiago Silva", "Player URL": "https://fbref.com/en/players/86e7deaf/scout/3243/Thiago-Silva-Scouting-Report"},
        {"Player Name": "Idrissa Gana Gueye", "Player URL": "https://fbref.com/en/players/72c812f3/scout/3243/Idrissa-Gana-Gueye-Scouting-Report"},
        {"Player Name": "Kylian Mbappé", "Player URL": "https://fbref.com/en/players/42fd9c7f/scout/3243/Kylian-Mbappe-Scouting-Report"},
        {"Player Name": "Juan Bernat", "Player URL": "https://fbref.com/en/players/433dc9a2/scout/3243/Juan-Bernat-Scouting-Report"},
        {"Player Name": "Marco Verratti", "Player URL": "https://fbref.com/en/players/1467af0d/scout/3243/Marco-Verratti-Scouting-Report"},
        {"Player Name": "Marquinhos", "Player URL": "https://fbref.com/en/players/d5f2f82b/scout/3243/Marquinhos-Scouting-Report"},
        {"Player Name": "Neymar", "Player URL": "https://fbref.com/en/players/69384e5d/scout/3243/Neymar-Scouting-Report"},
        {"Player Name": "Mauro Icardi", "Player URL": "https://fbref.com/en/players/43b78598/scout/3243/Mauro-Icardi-Scouting-Report"},
        {"Player Name": "Thomas Meunier", "Player URL": "https://fbref.com/en/players/e162b013/scout/3243/Thomas-Meunier-Scouting-Report"},
        {"Player Name": "Presnel Kimpembe", "Player URL": "https://fbref.com/en/players/aef6e670/scout/3243/Presnel-Kimpembe-Scouting-Report"},
        {"Player Name": "Abdou Diallo", "Player URL": "https://fbref.com/en/players/fc62f087/scout/3243/Abdou-Diallo-Scouting-Report-Scouting-Report"},
        {"Player Name": "Pablo Sarabia", "Player URL": "https://fbref.com/en/players/9744ff80/scout/3243/Pablo-Sarabia-Scouting-Report"},
        {"Player Name": "Layvin Kurzawa", "Player URL": "https://fbref.com/en/players/41e7ef1c/scout/3243/Layvin-Kurzawa-Scouting-Report"},
        {"Player Name": "Leandro Paredes", "Player URL": "https://fbref.com/en/players/dff153a4/scout/3243/Leandro-Paredes-Scouting-Report"},
        {"Player Name": "Julian Draxler", "Player URL": "https://fbref.com/en/players/1a992676/scout/3243/Julian-Draxler-Scouting-Report"},
        {"Player Name": "Colin Dagba", "Player URL": "https://fbref.com/en/players/29c02c9f/scout/3243/Colin-Dagba-Scouting-Report"},
        {"Player Name": "Edinson Cavani", "Player URL": "https://fbref.com/en/players/527f063d/scout/3243/Edinson-Cavani-Scouting-Report"},
        {"Player Name": "Ander Herrera", "Player URL": "https://fbref.com/en/players/436a9dd0/scout/3243/Ander-Herrera-Scouting-Report"},
        {"Player Name": "Thilo Kehrer", "Player URL": "https://fbref.com/en/players/51dbeea9/scout/3243/Thilo-Kehrer-Scouting-Report"},
        {"Player Name": "Eric Maxim Choupo-Moting", "Player URL": "https://fbref.com/en/players/0e6bc38c/scout/3243/Eric-Maxim-Choupo-Moting-Scouting-Report"},
        {"Player Name": "Tanguy Nianzou", "Player URL": "https://fbref.com/en/players/5853ac44/scout/3243/Tanguy-Nianzou-Scouting-Report"},
        {"Player Name": "Alphonse Areola", "Player URL": "https://fbref.com/en/players/2f965a72/scout/3243/Alphonse-Areola-Scouting-Report"},
        {"Player Name": "Sergio Rico", "Player URL": "https://fbref.com/en/players/5f0241b2/scout/3243/Sergio-Rico-Scouting-Report"},
        {"Player Name": "Mitchel Bakker", "Player URL": "https://fbref.com/en/players/a8567629/scout/3243/Mitchel-Bakker-Scouting-Report"},
        {"Player Name": "Marcin Bułka", "Player URL": "https://fbref.com/en/players/db87a2c4/scout/3243/Marcin-Bulka-Scouting-Report"},
        {"Player Name": "Adil Aouchiche", "Player URL": "https://fbref.com/en/players/b94b8be1/scout/3243/Adil-Aouchiche-Scouting-Report"},
        {"Player Name": "Loïc Mbe Soh", "Player URL": "https://fbref.com/en/players/8f534f8e/scout/3243/Loic-Mbe-Soh-Scouting-Report"},
        {"Player Name": "Arthur Zagre", "Player URL": "https://fbref.com/en/players/3225654e/scout/3243/Arthur-Zagre-Scouting-Report"},
        {"Player Name": "Jesé", "Player URL": "https://fbref.com/en/players/4a537185/scout/3243/Jese-Scouting-Report"},
        {"Player Name": "Garissone Innocent", "Player URL": "https://fbref.com/en/players/3e6f30a9/scout/3243/Garissone-Innocent-Scouting-Report"},
        {"Player Name": "Arnaud Kalimuendo", "Player URL": "https://fbref.com/en/players/bf4c41f9/scout/3243/Arnaud-Kalimuendo-Scouting-Report"}
    ]

    # Convertir en DataFrame
    players_df = pd.DataFrame(players_data)
    
    # Sauvegarder sous forme de fichier CSV temporaire
    temp_csv_path = "../../data/scouting_report_psg.csv"
    players_df.to_csv(temp_csv_path, index=False)

    # Initialiser l'extracteur de données
    extractor = DataPlayerExtractor(players_data_path=temp_csv_path)
    extractor.process_players()
```
